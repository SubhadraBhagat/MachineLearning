#Credit card fraud detection using logistic regression

This project aims to detect the fraudulent transaction which occurs while using credit card, The credit card transaction data is taken from kaggle https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud , The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions, so in order to make it balanced I have taken equivalent percentage of the legit transaction sample from the dataset, before feeding it to logistic regression model, which is a binary classification problem, beacuse in this project the primary goal is to classify the legit and fraudulent transaction. Also this data only contains numerical input variables which are the result of a PCA transformation, because credit card data are very confidential and sensitive to public exposure. After training the data using logistic regression model, it is found that the model is performing  well on both the training and test set with 94% and 91% of accuracy respectively. 

## Importing libraries 

```
import numpy as np                                   #numpy library useful for making numpy array
import pandas as pd                                  #to structure data in pandas dataframe
import matplotlib.pyplot as plt                      #useful for making plots and viusalization
```

```
dataset = pd.read_csv('creditcard.csv')              #uploading data in pandas dataframe
```

## Handling missing data

imputaion or drop, since data set is too big, dropping few data won't affect much, hence dropping the missing value rows

```
dataset =  dataset.dropna(how = 'any') 
```

```
dataset.isnull().sum()           #no missing data successfully dropped
```

```
dataset.shape                     #one row with missing value has been dropped 
```

```
dataset['Class'].value_counts()   #highly unbalanced data
```

## splitting legit and fraudulent transaction

```
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```
print(legit.shape)
print(fraud.shape)
```

```
#compare values for normal and fraululent classes, compare value for mean
dataset.groupby('Class').mean()            #wide diff b/w mean this id imp thats how ML predicts 
```

## Under Sampling 

```
#build sample dataset from original dataset containing similar distributuion of legit and fraudulent transactoion
#no of fraudulent transaction ---> 492
#goal is to make eqaul distribution so we will take 492 data from legit data to sample data

legit_sample = legit.sample(n = 492) #it will take 492 data point randomly from legit, good way to take data into sample
```

```
concateneted_data = pd.concat([legit_sample, fraud], axis = 0)  # axis = 0 adding row wise concatenation. 
                                                                # axis = 1 add column wise dont want that
```

```
concateneted_data.head()  #random value has taken form legit to legit_sample 
```

```
concateneted_data.shape
```

```
concateneted_data['Class'].value_counts()  #equal sample distribution 
```

```
concateneted_data.groupby('Class').mean()  #imp to check whether sample is good or not, the nature of dataset has not 
#changed, and diff is still there as earlier so now we can put this sample to train adnd Ml will give good result 
```

## Splitting dataset into feature matrix and output

```
X = concateneted_data.iloc[:, :-1].values   #feature matrix
y = concateneted_data.iloc[:, -1].values    #output
```

```
print(X)
```

```
print(y)
```

## Splitting data into training set and test set

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 2)  
#test_size = 0.2 means using 20% of the orizinal data as test set, random state train and test , it would be 
#3 if there is CV set may be, stratify - means in both the training and test set 0 and 1 will be evenly distributed,
#if not mention then it would be very uneven distribution 
#may be in test set there will be only fraudulent transaction and in training set only legit transaction
```

```
print(X.shape, X_train.shape, X_test.shape) 
```

## Training the Logistic Regression model on the Training set

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

## Model evaluation

```
#accuracy score
#accuracy on training data
from sklearn.metrics import accuracy_score     
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train) #X_train_prediction = h(x), y_train = y, 
                                                                     # accuracy = h(x) - y
```

```
print('accuracy on training data: ', training_data_accuracy)
#generally when prediction are more than 75% then predictions are good here in our case its 94% so our 
#predicted accuracy is very good
```

## Predicting test set results

```
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making confusion matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
#y_pred = h(x), y_test = y, accuracy = h(x) - y
```
